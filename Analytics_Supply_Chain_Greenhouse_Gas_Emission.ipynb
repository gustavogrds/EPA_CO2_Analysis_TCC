```python
df2 = pd.read_excel('/content/drive/MyDrive/SupplyChainEmissionFactorsforUSIndustriesCommodities.xlsx',sheet_name='2016_Summary_Commodity')

df2['ANO'] = 2016
df2.head()
```

```python
# Kmeans Ano2016

#EXCLUINDO TIPO STRING E NULL / PREPROCESSAMENTO
X = df2[['Supply Chain Emission Factors without Margins','Margins of Supply Chain Emission Factors','Supply Chain Emission Factors with Margins']]

from sklearn.preprocessing import scale

df_preprocessing = scale(X)

print(df_preprocessing)
# Declara o modelo
clf = KMeans(n_clusters = 3 , random_state= 1984)  # seed, para a reprodutibilidade dos resultados
# Calcula
clf.fit(X)
# Resultados
labels = clf.labels_
centroids = clf.cluster_centers_
print(labels)
print(centroids)

# seu código aqui

df2['cluster']=labels

df2.head()
```

```python
# DEFINIÇÂO NUMERO DE KLUSTERS

#EXCLUINDO TIPO STRING E NULL / PREPROCESSAMENTO
X = df.drop(columns=['Commodity Code','Commodity Name','Substance','Unit', 'Unnamed: 7','ANO'] )

from sklearn.preprocessing import scale

df_preprocessing = scale(X)

print(df_preprocessing)

```

```python
#REGRA DO COTOVELO ( DEF K=3 )
from sklearn.cluster import KMeans

inertia = []

for n in range(1 , 11):
    clf = KMeans(n_clusters = n , random_state= 1984)
    clf.fit(df_preprocessing)
    inertia.append(clf.inertia_)
plt.figure(1 , figsize = (15 ,7))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
```

```python
# Declara o modelo
clf = KMeans(n_clusters = 3 , random_state= 1984)  # seed, para a reprodutibilidade dos resultados
# Calcula
clf.fit(X)
3# Resultados
labels = clf.labels_
centroids = clf.cluster_centers_
print(labels)
print(centroids)

# seu código aqui

df['cluster_Total']=labels

df.head()
```

```python
df_cluster = df[df['cluster_Total'] == 1]
sn.barplot(y=df_cluster['Substance'], x=df_cluster['cluster_Total'], hue=df_cluster['Commodity Name'])
```
